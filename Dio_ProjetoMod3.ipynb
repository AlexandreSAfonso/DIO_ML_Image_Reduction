# Redução de Dimensionalidade em Imagens para Redes Neurais

##Objetivo
O principal objetivo desse projeto é implementar, do zero, funções para:

Converter imagens coloridas para tons de cinza: Esta conversão é feita utilizando uma fórmula ponderada que leva em consideração a luminância (brilho) da imagem.

Binarizar a imagem: Transformar a imagem para uma versão com apenas duas cores (preto e branco), usando um limiar de intensidade.

Essas transformações ajudam a reduzir a quantidade de informações que precisam ser processadas, mantendo o essencial para a análise.

Leitura de arquivo e armazenamento em memória (imagem_original)



```python
from google.colab import drive
from PIL import Image
from IPython.display import display

drive.mount('/content/drive')
caminho_imagem = "/content/drive/MyDrive/AleArthur.jpg"
imagem_original = Image.open(caminho_imagem)

#display(imagem_original)
```

Função para gerar imagem e escalas de Cinza

```python
def converter_para_cinza(imagem_original):
  largura, altura = imagem_original.size
  imagem_cinza = Image.new("L", (largura, altura))  # Escala de cinza

  for y in range(altura):
      for x in range(largura):
          r, g, b = imagem_original.getpixel((x, y))
          valor_cinza = int(0.299 * r + 0.587 * g + 0.114 * b)
          imagem_cinza.putpixel((x, y), valor_cinza)

  return imagem_cinza
```

Função para transformar a imagem em Preto e Branco apena, usando um limiar para segmentar os tons de cinza em preto ou branco.

```python
def binarizar_imagem(imagem_cinza, limiar=127):
  largura, altura = imagem_cinza.size
  imagem_binaria = Image.new("1", (largura, altura))  # Binária

  for y in range(altura):
      for x in range(largura):
          valor_cinza = imagem_cinza.getpixel((x, y))
          valor_binario = 1 if valor_cinza > limiar else 0
          imagem_binaria.putpixel((x, y), valor_binario)


  return imagem_binaria
```

Chamando as funções e exibindo resultados.

```python
# Converter a imagem para cinza
imagem_cinza = converter_para_cinza(imagem_original)

# Binarizar a imagem
imagem_binaria = binarizar_imagem(imagem_cinza)


# Exibir as imagens processadas
print("Imagem Original:")
display(imagem_original)

print("Imagem em Tons de Cinza:")
display(imagem_cinza)

print("Imagem Binarizada:")
display(imagem_binaria)
```
